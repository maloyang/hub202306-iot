# 太陽能資訊

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
base_path = './02_data/'
```

```python
df_temp = pd.read_csv(os.path.join(base_path, 'plant02_2023-05-01.csv'))
df_temp
```

```python
df1=df_temp.loc[df_temp['id'] == 1]
df2=df_temp.loc[df_temp['id'] == 2]
df3=df_temp.loc[df_temp['id'] == 3]
```

```python
df1.head()
```

```python
dtime_list = df1['time'].to_list()
total_e1 = df1['total_energy'].to_list()
total_e2 = df2['total_energy'].to_list()
total_e3 = df3['total_energy'].to_list()
```

```python
plt.subplots(figsize=(15,5))
plt.plot(dtime_list[::5], total_e1[::5])
plt.xlabel('time')
plt.ylabel('kwh')
plt.xticks(dtime_list[0::10], rotation=90)
plt.grid()
plt.show()
```

```python
plt.subplots(figsize=(15,5))
plt.plot(dtime_list[::5], total_e1[::5])
plt.plot(dtime_list[::5], total_e2[::5])
plt.plot(dtime_list[::5], total_e3[::5])
plt.xlabel('time')
plt.ylabel('kwh')
plt.xticks(dtime_list[0::10], rotation=90)
plt.grid()
plt.show()
```

```python
df1_tmp = df1.pivot_table(index='time',values='total_energy')
df1_tmp.index= pd.to_datetime(df1_tmp.index)
df1_tmp = df1_tmp.resample('H').last()
```

```python
df1_tmp['e_diff'] = df1_tmp['total_energy'].diff().fillna(0)
```

```python
df2_tmp = df2.pivot_table(index='time',values='total_energy')
df2_tmp.index= pd.to_datetime(df2_tmp.index)
df2_tmp = df2_tmp.resample('1H').last()
df2_tmp['e_diff'] = df2_tmp['total_energy'].diff().fillna(0)
df2_tmp
```

```python
df3_tmp = df3.pivot_table(index='time',values='total_energy')
df3_tmp.index= pd.to_datetime(df3_tmp.index)
df3_tmp = df3_tmp.resample('1H').last()
df3_tmp['e_diff'] = df3_tmp['total_energy'].diff().fillna(0)
df3_tmp
```

```python
dtime_diff = df1_tmp.index.to_list()
e1_diff = df1_tmp['e_diff'].to_list()
e2_diff = df2_tmp['e_diff'].to_list()
e3_diff = df3_tmp['e_diff'].to_list()
```

```python
dtime_diff
```

```python
plt.subplots(figsize=(15,5))
plt.plot(dtime_diff, e1_diff, label='e1')
plt.plot(dtime_diff, e2_diff, label='e2')
plt.plot(dtime_diff, e3_diff, label='e3')
plt.xlabel('time')
plt.ylabel('kwh')
plt.xticks(dtime_diff, rotation=90)
plt.legend()
plt.grid()
plt.show()
```
