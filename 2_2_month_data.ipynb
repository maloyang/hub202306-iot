## 1> import modules


```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

## 2> Data Reading

```python
base_path = './pm_data/'
```

```python
total_e_list = list()
df_temp = pd.read_csv(os.path.join(base_path, 'farm1_2023-04-01.csv'))
temp_list = df_temp['total_e'].to_list()
total_e_list.append(temp_list[0])
```

```python
total_e_list
```

```python
for file in os.listdir(base_path):
    df_temp = pd.read_csv(os.path.join(base_path, file))
    temp_list = df_temp['total_e'].to_list()
    total_e_list.append(temp_list[-1])
    
```

```python
len(total_e_list)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 31), total_e_list[1:])
plt.xlabel('time')
plt.ylabel('total_e')
plt.grid()
plt.show()
```

```python
e_day_list = list()
for i in range(len(total_e_list)-1):
    e = total_e_list[i+1] - total_e_list[i]
    e_day_list.append(e)
print(e_day_list)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 31), e_day_list)
plt.xlabel('time')
plt.ylabel('e')
plt.grid()
plt.show()
```
